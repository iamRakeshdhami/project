# FILLING THE NULL VALUES

```python
Fifa['Loaned From'].fillna('No',inplace=True)
```

# for column with dtype object, filling NaN values with mode and for column with dtype float, filling NaN values with ceil.mean for more accurate answer

```python
for column in Fifa:
    if Fifa[column].isnull().any():
        if Fifa[column].dtypes==object:
            Fifa[column].fillna(Fifa[column].mode()[0],inplace=True)
            print('{0} -- {1}'.format(column, Fifa[column].isnull().sum()))
        elif Fifa[column].dtypes==float:
            Fifa[column].fillna(math.ceil(Fifa[column].mean()),inplace=True)
            print('{0} -- {1}'.format(column, Fifa[column].isnull().sum()))
        else:
            pass
```

# Cleaning the wage column

```python
def extract_value_from(Value):
    out = Value.replace('€', '')
    if 'M' in out:
        out = float(out.replace('M', ''))*1000000
    elif 'K' in Value:
        out = float(out.replace('K', ''))*1000
    return float(out)
```

```python
df['Value'] = df['Value'].apply(lambda x: extract_value_from(x))
df['Wage'] = df['Wage'].apply(lambda x: extract_value_from(x))

df['Wage'].head()
```

```python
sns.set(style = 'darkgrid')
plt.figure(figsize = (10,4))
sns.barplot(x=df["Wage"], data=df, color = 'g')
plt.xlabel('Wage Range for Players')
plt.ylabel('Count of the Players')
plt.title('Comparing the wages of players', fontsize=20)
plt.show()
```
